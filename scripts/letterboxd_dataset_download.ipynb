```python
import pandas as pd
import os 

# Local path for your Parquet snapshot
parquet_path = "letterboxd_all_movies.parquet"

# If local copy exists, just load it
if os.path.exists(parquet_path):
    df = pd.read_parquet(parquet_path)
else:
    # Otherwise, stream the HF Parquet version
    df = pd.read_parquet(
        "hf://datasets/pkchwy/letterboxd-all-movie-data@refs/convert/parquet"
    )
    # Save locally for fast reruns and reproducibility
    df.to_parquet(parquet_path)
    
# df is ready for analysis
print(df.shape)
print(df.columns)
```
